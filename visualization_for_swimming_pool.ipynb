```
df2 = df.filter(['영업상태명', '도로명전체주소','사업장명', '좌표정보(X)','좌표정보(Y)'])
df2 = df2[df2['영업상태명']=='영업/정상']
```

```
waterpark = df2[df2['사업장명'].str.contains("워터|물|오션|스파|아쿠아|온천|리조트")].dropna()
waterpark = waterpark.drop_duplicates(['사업장명'], keep='first')
waterpark.columns
```

```
s=pd.read_excel('/content/drive/My Drive/서울특별시 수영장 정보_20180521.xlsx',encoding='cp949')
seoul = s.filter(['상세영업상태명', '도로명전체주소','사업장명', '위치정보(X)','위치정보(Y)']).dropna()
seoul = seoul[seoul['상세영업상태명']=='영업중']
seoul.columns = ['영업상태명', '도로명전체주소', '사업장명', '좌표정보(X)', '좌표정보(Y)']
```

```
all = pd.concat([waterpark,seoul],ignore_index=True)
all.to_excel('all_waterpark.xlsx')
```

```
from pyproj import Proj, transform
```

```
# 위경도 변환 함수
proj_UTMK = Proj(init='epsg:2097')
proj_WGS84 = Proj(init='epsg:4019')
def transform_utmk_to_w84(df):
    return pd.Series(transform(proj_UTMK, proj_WGS84, df['좌표정보(X)'], df['좌표정보(Y)']), index=['x', 'y'])
```

```
# 좌표 위경도 변환 확인 

from pyproj import Proj, transform

# inProj = Proj(init='epsg:5173')
inProj = Proj(init='epsg:2097')
# outProj = Proj(init='epsg:4326')
outProj = Proj(init='epsg:4019')
x1,y1 = 212731.574257,450534.170653
x2,y2 = transform(inProj,outProj,x1,y1)
print(x2,y2)
# 서울특별시 강동구 양재대로 1665 (명일동) 37.5542,127.1437664
```

```
all[['경도','위도']] = all.apply(transform_utmk_to_w84,axis=1)
```

```
all
```

```
map_osm = folium.Map(location=[37.5103, 126.982], zoom_start=12)
```

```
for i in range(len(all.index)):
  name = all.loc[i, '사업장명']
  lat = all.loc[i, '위도']
  lng = all.loc[i, '경도']
    
    # 추출한 정보를 지도에 표시
  marker = folium.Marker([lat,lng], popup=name,icon=folium.Icon(color='red'))
  marker.add_to(map_osm)

map_osm.save('물놀이_시설_분포.html')
map_osm
```
